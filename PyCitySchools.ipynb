```python
#importing the necessary libraries
import pandas as pd

```

```python
# loading dataset needed for analysis
schools = pd.read_csv("schools_complete.csv")
students = pd.read_csv("students_complete.csv")

# merging the two datasets for district analysis
school_data = pd.merge(students, schools, how='left', on="school_name")

```

```python
school_data.head()
```

District Summary

•	Create a high level snapshot (in table form) of the district's key metrics, including:

•	Calculate the total number of schools

•	Calculate the total number of students

•	Calculate the total budget

•	Calculate the average math score

•	Calculate the average reading score

•	Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

•	Calculate the percentage of students with a passing math score (70 or greater)

•	Calculate the percentage of students with a passing reading score (70 or greater)

•	Create a dataframe to hold the above results

•	Optional: give the displayed data cleaner formatting


```python
# total number of schools
tot_schools = school_data['school_name'].nunique()

# total number of students
tot_students = school_data["Student ID"].nunique()

# total budget
tot_budget = schools['budget'].sum()

# average math score
avg_math = school_data['math_score'].mean()

# average reading score
avg_read = school_data['reading_score'].mean()

# overall passing rate
ovr_pass = (avg_math + avg_read)/2

# Percentage of students with a passing math score (70 or greater)
pass_math = school_data[school_data["math_score"]>=70]
percent_pass = (len(pass_math)/len(school_data['Student ID']))*100

# Percentage of students with a passing reading score (70 or greater)
pass_read = school_data[school_data["reading_score"]>=70]
percent_read = len(pass_read)/len(school_data['Student ID'])*100

# Key metrics
district = [{"Total number of schools": tot_schools, 
            "Total number of students" : tot_students, 
            "Total Budget" : tot_budget,
            "Average math score" : round(avg_math,2), 
            "Average reading score" : round(avg_read,2), 
            "Overall passing rate" : round(ovr_pass,2), 
            "% Passing math score" : percent_pass, 
            "% Passing reading score" : percent_read}]

district_summary = pd.DataFrame(district)

# Formatting
district_summary['Total Students'] = district_summary["Total number of students"].map("{:,}".format)
district_summary['Total Budget'] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary['% Passing math score'] = district_summary["% Passing math score"].map("{:,.2f}%".format)
district_summary['% Passing reading score'] = district_summary["% Passing reading score"].map("{:,.2f}%".format)

district_summary
```

School Summary

•	Create an overview table that summarizes key metrics about each school, including:

o	School Name

o	School Type

o	Total Students

o	Total School Budget

o	Per Student Budget

o	Average Math Score

o	Average Reading Score

o	% Passing Math (The percentage of students that passed math.)

o	% Passing Reading (The percentage of students that passed reading.)

o	% Overall Passing (The percentage of students that passed math and reading.)


```python
# School Type
sch_type = schools.set_index(["school_name"])["type"]

# Total students per school
sch_students = schools.set_index(["school_name"])["size"]

# Total School Budget
sch_budget = schools.set_index(["school_name"])["budget"]

# Per Student Budget
per_stud = sch_budget/sch_students

# Average Math Score Per Schoool
per_avg_math = school_data.groupby('school_name')["math_score"].mean()

# Average Reading Score Per School
per_avg_read = school_data.groupby('school_name')["reading_score"].mean()

# % Passing Math Per school
per_pass_math = school_data[school_data['math_score'] >=70]
per_sch_math = ((per_pass_math.groupby('school_name').size())/sch_students)*100

# % Passing Reading Per School
per_pass_read = school_data[school_data['reading_score'] >=70]
per_sch_read = ((per_pass_read.groupby('school_name').size())/sch_students)*100

# % Overall Passing Per School
per_ovr_pass = (per_sch_read + per_sch_math)/2

# Overview table
school_summary  = pd.DataFrame({'School Type': sch_type,
                    "Total Students": sch_students,
                    "Total School Budget": sch_budget,
                    "Per Student Budget": per_stud,
                    "Average Math Score": round(per_avg_math,2),
                    "Average Reading Score": round(per_avg_read,2),
                    "% Passing Math": per_sch_math,
                    "% Passing Reading": per_sch_read,
                    "% Overall Passing": per_ovr_pass})

# formatting
school_summary['Total Students'] = school_summary["Total Students"].map("{:,}".format)
school_summary['Total School Budget'] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary['Per Student Budget'] = school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary['% Passing Math'] = school_summary["% Passing Math"].map("{:,.2f}".format)
school_summary['% Passing Reading'] = school_summary["% Passing Reading"].map("{:,.2f}".format)
school_summary['% Overall Passing'] = school_summary["% Overall Passing"].map("{:,.2f}".format)

school_summary
```

Math Scores by Grade**

•	Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.


```python
# Categorizing by Grades and getting the average math scores

nine_grade = school_data[school_data['grade']== '9th']
ten_grade = school_data[school_data['grade']== '10th']
elev_grade = school_data[school_data['grade']== '11th']
twel_grade = school_data[school_data['grade']== '12th']

avg_nine = nine_grade.groupby('school_name')['math_score'].mean()
avg_ten = ten_grade.groupby('school_name')['math_score'].mean()
avg_elev = elev_grade.groupby('school_name')['math_score'].mean()
avg_twel = twel_grade.groupby('school_name')['math_score'].mean()

math_by_grade = pd.DataFrame({"9th Grade" : round(avg_nine,2),
                              "10th Grade": round(avg_ten,2),
                              "11th Grade": round(avg_elev,2),
                              "12th Grade": round(avg_twel,2)})
math_by_grade

```

Scores by School Spending

•	Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:

o	Average Math Score

o	Average Reading Score

o	% Passing Math (The percentage of students that passed math.)

o	% Passing Reading (The percentage of students that passed reading.)

o	% Overall Passing (The percentage of students that passed math and reading.)


```python
# Setting the bins
spend_bin = [0, 580, 620, 650, 700]
spend_labels = ["<$580", "$580-$620", "$620-$650", "$650-$700"]

sch_spend = school_summary.copy()

# Converting per Student Budget to float and replacing $ and commas if present
sch_spend['Per Student Budget'] = sch_spend['Per Student Budget'].replace({'\\$': '', ',': ''}, regex=True).astype(float)

# Applying spending bins and labels present above
sch_spend['Average Spending Ranges'] = pd.cut(sch_spend['Per Student Budget'], bins=spend_bin, labels=spend_labels)

# Converting average spend column to string
sch_spend['Average Spending Ranges'] = sch_spend['Average Spending Ranges'].astype(str)

# Formatting per student budget column
sch_spend['Per Student Budget'] = sch_spend["Per Student Budget"].map("${:,.2f}".format)
sch_spend
```

```python
# Converting below columns back to numeric to get mean
sch_spend['% Passing Math'] = pd.to_numeric(sch_spend['% Passing Math'], errors='coerce')
sch_spend['% Passing Reading'] = pd.to_numeric(sch_spend['% Passing Reading'], errors='coerce')
sch_spend['% Overall Passing'] = pd.to_numeric(sch_spend['% Overall Passing'], errors='coerce')

avg_math_spend = sch_spend.groupby('Average Spending Ranges')['Average Math Score'].mean()
avg_read_spend = sch_spend.groupby('Average Spending Ranges')['Average Reading Score'].mean()
percent_math_spend = sch_spend.groupby('Average Spending Ranges')['% Passing Math'].mean()
percent_read_spend = sch_spend.groupby('Average Spending Ranges')['% Passing Reading'].mean()
percent_ovr_spend = sch_spend.groupby('Average Spending Ranges')['% Overall Passing'].mean()

spend_summary = pd.DataFrame({"Average Math Score": avg_math_spend,
                              "Average Reading Score": avg_read_spend,
                              "% Passing Math": percent_read_spend,
                              "% Passing Reading": percent_math_spend,
                              "% Overall Passing": percent_ovr_spend})

# Formatting
spend_summary['% Passing Math'] = spend_summary["% Passing Math"].map("{:,.2f}%".format)
spend_summary['% Passing Reading'] = spend_summary["% Passing Reading"].map("{:,.2f}%".format)
spend_summary['% Overall Passing'] = spend_summary["% Overall Passing"].map("{:,.2f}%".format)

spend_summary
```

Scores by School Size

•	Repeat the above breakdown, but this time group schools based on a reasonable approximation of school size (Small, Medium, Large).


```python
# Setting bins 
size_bin = (0, 1000, 3000, 5000)
size_labels = ("Small (<1000)", "Medium (1000-3000)", "Large (3000-5000)")

sch_size = school_summary.copy()

# # Converting Total Students to int
sch_size['Total Students'] = sch_size['Total Students'].replace({',': ''}, regex=True).astype(int)

# Applying size bins and labels present above
sch_size['School Size'] = pd.cut(sch_size['Total Students'], bins=size_bin, labels=size_labels)

# Converting Scores by School Size to string
sch_size['School Size'] = sch_size['School Size'].astype(str)

# Formatting total students
sch_size['Total Students'] = sch_size["Total Students"].map("{:,}".format)

sch_size



```

```python
# Converting below columns back to numeric to get mean
sch_size['% Passing Math'] = pd.to_numeric(sch_size['% Passing Math'], errors='coerce')
sch_size['% Passing Reading'] = pd.to_numeric(sch_size['% Passing Reading'], errors='coerce')
sch_size['% Overall Passing'] = pd.to_numeric(sch_size['% Overall Passing'], errors='coerce')

avg_math_size = sch_size.groupby('School Size')['Average Math Score'].mean()
avg_read_size = sch_size.groupby('School Size')['Average Reading Score'].mean()
percent_math_size = sch_size.groupby('School Size')['% Passing Math'].mean()
percent_read_size = sch_size.groupby('School Size')['% Passing Reading'].mean()
percent_ovr_size = sch_size.groupby('School Size')['% Overall Passing'].mean()

size_summary = pd.DataFrame({"Average Math Score": avg_math_size,
                              "Average Reading Score": avg_read_size,
                              "% Passing Math": percent_read_size,
                              "% Passing Reading": percent_math_size,
                              "% Overall Passing": percent_ovr_size})

# Formatting
size_summary['% Passing Math'] = size_summary["% Passing Math"].map("{:,.2f}%".format)
size_summary['% Passing Reading'] = size_summary["% Passing Reading"].map("{:,.2f}%".format)
size_summary['% Overall Passing'] = size_summary["% Overall Passing"].map("{:,.2f}%".format)

size_summary
```
